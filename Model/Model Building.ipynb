```python
# Loading the dataset
import pandas as pd

url = "https://raw.githubusercontent.com/dineshpiyasamara/LaptopPricePredictor/master/model%20building/laptop_price.csv"
data = pd.read_csv(url,encoding='latin-1')
data.head()
```

### Preprocessing

```python
data.shape # Checking the dimension
```

```python
data.isnull().sum() # Check whether missiong values exist
```

```python
data.info() # Some information about variables
```

```python
data.head(2)
```

```python
# To covert variables numeric, remove the srting part in the values of variable 'Ram' and 'Weight'
data['Ram'] = data['Ram'].astype(str).str.replace('GB','').astype('int32')
data['Weight'] = data['Weight'].astype(str).str.replace('kg','').astype('float32')
```

```python
data.head()
```

```python
data.info() # Check whether those variables' data type have changed
```

```python
# Calculate the correlation between 'Price_euros' and other numerical variables. after that we can decide less correlated columns from the dataset
data.select_dtypes(include='number').corr()['Price_euros']
```

```python
# See how many value counts in the 'Company' variable
data['Company'].value_counts()
```

```python
# Keep more frequent values same and change other values to 'Other'
def add_company(value):
    if value in ['Samsung','Mediacom','Razer','Microsoft','Vero','Xiaomi','Chuwi','Fujitsu','Google','LG','Huawei']:
        return 'Other'
    else:
        return value
data['Company'] = data['Company'].apply(add_company)
```

```python
# Check whether the change is applied
data['Company'].value_counts()
```

```python
# See how many value counts in the 'Product' variable
print(data['Product'].value_counts())
len(data['Product'].value_counts())
```

```python
# See how many value counts in the 'ScreenResolution' variable
data['ScreenResolution'].value_counts()
```

```python
# Create 2 new columns to indicate that where it is 'touchscreen' or 'ips' as 1 & 0 
data['TouchScreen'] = data['ScreenResolution'].apply(lambda x:1 if 'Touchscreen' in x else 0)
data['IPS'] = data['ScreenResolution'].apply(lambda x:1 if 'IPS' in x else 0)
```

```python
data.head()
```

```python
data['Cpu'].value_counts()
```

```python
# Extract 1st 3 words in the values in the 'Cpu' colum
data['Cpu_name'] = data['Cpu'].apply(lambda x:' '.join(x.split()[0:3]))
data['Cpu_name'].value_counts()
```

```python
# Keep more frequent values same and change other values to 'Other'
def set_processor(name):
    if name in ['Intel Core i7','Intel Core i5','Intel Core i3']:
        return name
    else:
        if name.split()[0]=='AMD':
            return 'AMD'
        else:
            return 'Other'
data['Cpu_name'] = data['Cpu_name'].apply(set_processor)
```

```python
data['Cpu_name'].value_counts()
```

```python
# see the 'Gpa' value counts
data['Gpu'].value_counts()
```

```python
# Extract 1st  words in the values in the 'Gpu' column
data['Gpu_name'] = data['Gpu'].apply(lambda x:' '.join(x.split()[0:1]))
data['Gpu_name'].value_counts()
```

```python
data.shape
```

```python
# Remove the record that 'Gpu_name' = 'ARM'
data = data[data['Gpu_name']!='ARM']
data.shape
```

```python
data.head()
```

```python
data['OpSys'].value_counts()
```

```python
# Keep more frequent values same and change other values to 'Other'
def set_os(name):
    if name in ['Windows 10','Windows 7','Windows 10 S']:
        return 'Windows'
    else:
        if name in ['macOS','Mac OS X']:
            return 'Mac'
        else:
            if name== 'Linux':
                return name
            else:
                return 'Other'
data['OpSys'] = data['OpSys'].apply(set_os)
```

```python
data['OpSys'].value_counts()
```

```python
data.head()
```

```python
# Drop the unwanted columns from the dataset
data = data.drop(columns=['laptop_ID','Inches','Product','Cpu','Gpu','ScreenResolution'])
```

```python
data.head()
```

```python
# Convert categorical columns to numerical ones
data = pd.get_dummies(data)
data.head()
```

```python
data.shape
```

### Building the model

```python
# Create X and y variable for building the model
X = data.drop('Price_euros',axis=1)
y = data['Price_euros']
```

```python
#!pip install scikit-learn
```

```python
# Split the dataset
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=123)
```

```python
X_train.shape , X_test.shape
```

```python
# Define a function to train the model and print the accuracy
def model_acc(model):
    model.fit(X_train,y_train)
    acc = model.score(X_test,y_test)
    print(str(model)+ '--> '+str(acc))
```

```python
# Fit different models and print the accuracy
from sklearn.linear_model import LinearRegression
LR = LinearRegression()
model_acc(LR)

from sklearn.linear_model import Lasso
Lasso = Lasso()
model_acc(Lasso)

from sklearn.tree import DecisionTreeRegressor
DT = DecisionTreeRegressor()
model_acc(DT)

from sklearn.ensemble import RandomForestRegressor
RF = RandomForestRegressor()
model_acc(RF)
```

```python
# Hyper parameter tuning
from sklearn.model_selection import GridSearchCV
parameters = {'n_estimators':[10,50,100],
             'criterion':['squared_error','absolute_error','poisson']}

grid_obj = GridSearchCV(estimator=RF,param_grid=parameters)

grid_fit = grid_obj.fit(X_train,y_train)

best_model = grid_fit.best_estimator_
best_model
```

```python
best_model.score(X_test,y_test) # accuracy of the best model
```
